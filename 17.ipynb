## Line plot

```python
import matplotlib.pyplot as plt

# Days and corresponding temperatures
days = [1, 2, 3, 4, 5]
temperatures = [25, 28, 22, 24, 30]

plt.gca().set_facecolor('#212121')  # Plot bg
plt.gcf().set_facecolor('black')  # Figure bg

plt.plot(days, temperatures, marker='o', 
         color='skyblue', linestyle='dashed')

plt.title("Daily Temperatures", color="w")
plt.xlabel("Day", color="white")
plt.ylabel("Temperature (Â°C)", color="#ffffff")
 
plt.grid(True) # If you prefer gridlines
plt.tick_params(colors='white')
plt.show()
```

## Bar Plot

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C']
sales = [120, 90, 300]

# Set background colors
plt.gca().set_facecolor('#212121')
plt.gcf().set_facecolor('black')

# Create bar plot
bars = plt.bar(products, sales, color='teal')

plt.title("Sales by Product Line", color='w')
plt.xlabel("Products", color='w')
plt.ylabel("Sales (Units)", color='w')

# Add values halfway on each bar
for bar in bars:
    yval = bar.get_height()  # Get the height (value) of each bar
    plt.text(bar.get_x() + bar.get_width()/2, yval / 2, int(yval), 
             ha='center', va='center', color='black', fontweight='bold')

plt.grid(False)
plt.tick_params(colors='white')
plt.show()
```

```python
import numpy as np

# Generate random data with a clear trend
np.random.seed(0)  # For reproducibility
num_points = 200

# Generate marketing spend data
marketing_spend = np.random.uniform(100, 1000, num_points)

# Generate sales data with a linear relationship plus some noise
slope = 0.1
intercept = 50
noise = np.random.normal(0, 20, num_points)  # Add some noise to make it realistic
sales = slope * marketing_spend + intercept + noise
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Create scatter plot
plt.scatter(marketing_spend, sales, 
            color='coral', alpha=0.6, 
            edgecolors='b', s=100)

# Fit and plot a trend line
m, b = np.polyfit(marketing_spend, sales, 1) 
plt.plot(marketing_spend, m * marketing_spend + b, 
         color='blue', linestyle='--')

plt.title("Marketing Spend vs Sales")
plt.xlabel("Marketing Spend ($)")
plt.ylabel("Sales ($)")

plt.show()
```
