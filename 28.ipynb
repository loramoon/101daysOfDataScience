## Scatter Plots

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {'Age': [22, 25, 29, 30, 35, 40, 50, 60, 70],
        'Income': [30000, 35000, 40000, 45000, 
                   60000, 80000, 90000, 120000, 150000]}
df = pd.DataFrame(data)

# Creating a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Income'], color='blue')
plt.title('Scatter Plot of Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(False) 
plt.show()
```

## Cross-tabulation

```python
# Sample categorical data
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
        'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']}
df_crosstab = pd.DataFrame(data)

# Creating a cross-tabulation
crosstab_result = pd.crosstab(df_crosstab['Gender'], df_crosstab['Purchased'])
print(crosstab_result)
```
