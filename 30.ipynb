## Correlation 

```python
import pandas as pd
import numpy as np

# Creating a sample dataset
data = {'Variable_A': [10, 20, 30, 40, 50],
        'Variable_B': [15, 30, 25, 35, 60]}
df = pd.DataFrame(data)

# Calculating the correlation
correlation = df.corr()
correlation
```

```python
import matplotlib.pyplot as plt

# Plotting the two variables to visualize their relationship
plt.figure(figsize=(6, 4))
plt.scatter(df['Variable_A'], df['Variable_B'], color='cyan')
plt.title('Scatter plot of Variable A vs. Variable B', color='white')
plt.xlabel('Variable A', color='white')
plt.ylabel('Variable B', color='white')

# Customizing plot background
plt.gca().set_facecolor('#111111')
plt.gcf().set_facecolor('#111111')
plt.tick_params(colors='white')

plt.show()
```

## Correlation Doesn’t Imply Causation

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated data
np.random.seed(0)
summer_months = np.linspace(1, 12, 100) 
ice_cream_sales = 100 * np.sin(summer_months / 12 * 2 * np.pi) + np.random.normal(0, 10, 100) + 200 
drowning_incidents = 50 * np.sin(summer_months / 12 * 2 * np.pi) + np.random.normal(0, 5, 100) + 100  

# Plotting
plt.scatter(ice_cream_sales, drowning_incidents, color='blue', alpha=0.6)
plt.title('Correlation between Ice Cream Sales and Drowning Incidents', color="w")
plt.xlabel('Ice Cream Sales', color="w")
plt.ylabel('Drowning Incidents', color="w")
plt.grid(True)  
plt.gcf().set_facecolor('#111111') 
plt.gca().set_facecolor('#111111') 
plt.tick_params(axis='both', colors='white')
plt.show()
```
