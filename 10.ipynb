## Scenario:

### Imagine we have two flavors of ice cream: chocolate and vanilla. We'll estimate the probability of someone choosing chocolate based on past choices.

### Frequentist Approach

```python
import numpy as np

# Simulate ice cream choices: 1 for chocolate, 0 for vanilla
np.random.seed(0)
choices = np.random.binomial(1, 0.6, 100)  # Assume 60% chance of choosing chocolate

probability_chocolate = np.mean(choices)

print(f"Frequentist estimate of probability of choosing chocolate: {probability_chocolate:.2f}")
```

### Bayesian Approach

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Observed data
chocolate_choices = 60
total_choices = 100
vanilla_choices = total_choices - chocolate_choices

# Define prior parameters for Beta distribution
a_prior, b_prior = 1, 1

a_posterior = a_prior + chocolate_choices
b_posterior = b_prior + vanilla_choices

# Calculate the mean of the posterior distribution
posterior_mean = a_posterior / (a_posterior + b_posterior)

print(f"Bayesian estimate of probability of choosing chocolate: {posterior_mean:.2f}")
```

```python

# Plot the prior and posterior distributions
x = np.linspace(0, 1, 100)
prior = beta.pdf(x, a_prior, b_prior)
posterior = beta.pdf(x, a_posterior, b_posterior)

plt.figure(figsize=(12, 6))
plt.plot(x, prior, label='Prior: Beta(1,1)', color='blue', linestyle='--')
plt.plot(x, posterior, label='Posterior: Beta(61,41)', color='red')
plt.fill_between(x, 0, posterior, color='red', alpha=0.2)

plt.gca().set_facecolor('#212121')  # Plot background
plt.gcf().set_facecolor('#212121')  # Figure background

plt.title('Bayesian Update of the Probability of Choosing Chocolate', color = "white")
plt.xlabel('Probability of Choosing Chocolate', color = "white")
plt.ylabel('Density', color = "white")

plt.legend()
plt.grid(False)
plt.tick_params(colors='white')
plt.show()
```
