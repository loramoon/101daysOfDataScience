## Histograms

```python
import matplotlib.pyplot as plt

# Sample data
data = [23, 45, 56, 22, 23, 56, 78, 45, 34, 56, 56, 67, 89, 23, 23]

# Creating a histogram
plt.hist(data, bins=10, color='skyblue')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Summary Statistics

```python
import numpy as np
import pandas as pd

# Converting data to a NumPy array
data = np.array([23, 45, 56, 22, 23, 56, 78, 45, 34, 56, 56, 67, 89, 23, 23])

# Calculating summary statistics
mean_value = np.mean(data)
median_value = np.median(data)
std_deviation = np.std(data)

print(f'Mean: {mean_value}')
print(f'Median: {median_value}')
print(f'Standard Deviation: {std_deviation}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.array([23, 45, 56, 22, 23, 56, 78, 45, 34, 56, 56, 67, 89, 23, 23])

# Calculating summary statistics
mean_value = np.mean(data)
median_value = np.median(data)
std_deviation = np.std(data)

# Creating the plot
plt.figure(figsize=(8, 6))

# Histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Data with Summary Statistics', color='white')
plt.xlabel('Value', color='white')
plt.ylabel('Frequency', color='white')

# Adding vertical lines for mean, median, and standard deviation
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='yellow', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(mean_value + std_deviation, color='green', linestyle='dashed', linewidth=2, label=f'Std Dev: {mean_value + std_deviation:.2f}')
plt.axvline(mean_value - std_deviation, color='green', linestyle='dashed', linewidth=2)

# Adding a legend
plt.legend()

# Customizing the plot's background and text colors to match your preferences
plt.gca().set_facecolor('#111111')
plt.gcf().set_facecolor('#111111')
plt.tick_params(colors='white')

plt.show()
```
