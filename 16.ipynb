## Identifying Patterns and Trends

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated temperature data for 365 days
days = np.arange(1, 366)

# Seasonal temperature changes
temperature = 10 + 10 * np.sin(2 * np.pi * days / 365)  

# Create the line plot
plt.gca().set_facecolor('#212121')  # Plot background
plt.gcf().set_facecolor('#212121')  # Figure background
plt.plot(days, temperature, color='red', label='Daily temperature')

plt.title('Temperature 2024', color = "white")
plt.xlabel('Day of the Year', color="white")
plt.ylabel('Temperature C', color="white")

plt.legend()
plt.grid(True)
plt.tick_params(colors='white')
plt.show()
```

## Simplifying Complex Data

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated advertising spend and sales data
np.random.seed(0)
ad_spend = np.random.rand(100) * 1000 
# adding some noise
sales = ad_spend * 1.5 + np.random.randn(100) * 100  


plt.gca().set_facecolor('#212121')  
plt.gcf().set_facecolor('#212121')  
plt.scatter(ad_spend, sales, color='lime', label='Sales vs Ad Spend ($)')
plt.title('Advertising Spend vs Sales ($)', color='white')
plt.xlabel('Ad Spend ($)', color='white')
plt.ylabel('Sales ($)', color='white')
plt.legend(facecolor='#212121')

plt.legend()
plt.grid(True)
plt.tick_params(colors='white')
plt.show()
```

## Enhanced Communication

```python
# Market share data for five products
products = ['Product A', 
            'Product B', 
            'Product C', 
            'Product D', 
            'Product E']
market_share = [35, 25, 15, 15, 10]

def format_autopct(pct):
    return f'{pct:.1f}%'

plt.figure(facecolor='#212121')
wedges, texts, autotexts = plt.pie(market_share, labels=products, autopct=format_autopct, 
                                   colors=['cyan', 'lime', 'orange', 'magenta', 'yellow'],
                                   textprops={'color': 'white'})
for autotext in autotexts:
    autotext.set_color('black')

plt.title('Market Share of Products', color='white')
plt.show()
```
