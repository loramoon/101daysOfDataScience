## Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd

# Load sample data
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)

# Apply PCA, reducing to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print(f'Explained variance ratios: {pca.explained_variance_ratio_}')
```

```python
import matplotlib.pyplot as plt

y = data.target

# Create a scatter plot of the PCA results
plt.figure(figsize=(8, 6))
for target, color in zip(range(3), ['r', 'g', 'b']):
    plt.scatter(X_pca[y == target, 0], X_pca[y == target, 1], 
                label=data.target_names[target], color=color)

plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()
plt.show()
```

## t-SNE (t-Distributed Stochastic Neighbor Embedding)

```python
from sklearn.manifold import TSNE

# Apply t-SNE to reduce dimensions to 2 for visualization
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

print(f't-SNE output shape: {X_tsne.shape}')
```

```python
y = data.target

# Create a scatter plot of the t-SNE results
plt.figure(figsize=(8, 6))
for target, color in zip(range(3), ['r', 'g', 'b']):
    plt.scatter(X_tsne[y == target, 0], X_tsne[y == target, 1], 
                label=data.target_names[target], color=color)

plt.title('t-SNE Visualization of the Iris Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid()
plt.show()
```

## Linear Discriminant Analysis (LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Apply LDA, using the class labels
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, data.target)

print(f'LDA output shape: {X_lda.shape}')
```

```python
# Create a scatter plot of the LDA results
plt.figure(figsize=(8, 6))
for target, color in zip(range(3), ['r', 'g', 'b']):
    plt.scatter(X_lda[y == target, 0], X_lda[y == target, 1], 
                label=data.target_names[target], color=color)

plt.title('LDA of Iris Dataset')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend()
plt.grid()
plt.show()
```
