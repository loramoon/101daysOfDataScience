## Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 0
std_dev = 1
size = 1000

# Generate data
data = np.random.normal(mean, std_dev, size)

# Plot the histogram
plt.figure(figsize=(10, 8))  # Adjust figure height (width, height)

# Set black background for the plot and figure
plt.gca().set_facecolor('#212121')  # Plot background
plt.gcf().set_facecolor('#212121')  # Figure background

# Plot histogram
count, bins, ignored = plt.hist(data, bins=100, density=True, alpha=0.6, color='g')

# Plot the bell curve (PDF of the normal distribution)
x = np.linspace(min(bins), max(bins), 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'w', linewidth=2)  # Use white for the curve to contrast against the black background

# Set y-axis limits to make the plot taller
plt.ylim(0, 0.5)  # Increase the upper limit to make the plot appear taller

# Add a title and labels with white text to contrast against the black background
plt.title('Normal Distribution with Bell Curve', color='white')
plt.xlabel('Value', color='white')
plt.ylabel('Density', color='white')

# Set tick colors to white
plt.tick_params(colors='white')

plt.show()
```

## Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_trials = 10
p_success = 0.5
size = 1000

# Generate data
data = np.random.binomial(n_trials, p_success, size)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=range(n_trials + 2), density=True, alpha=0.6, color='g')

plt.gca().set_facecolor('#212121')  # Plot background
plt.gcf().set_facecolor('#212121')  # Figure background

# Add a title and labels
plt.title('Binomial Distribution',color='white')
plt.xlabel('Number of Successes', color='white')
plt.ylabel('Frequency', color='white')
plt.tick_params(colors='white')
plt.show()
```

## Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
mean_impacts_per_year = 5  # Average number of meteorite impacts per year
size = 1000                # Number of simulations

# Generate data using the Poisson distribution
data = np.random.poisson(mean_impacts_per_year, size)

# Plot
plt.figure(figsize=(10, 6))

# Plot histogram
plt.hist(data, bins=range(0, max(data) + 1), density=True, alpha=0.6, color='g')

plt.title('Number of Meteorite Impacts per Year', color='white')
plt.xlabel('Number of Impacts', color='white')
plt.ylabel('Frequency', color='white')

plt.gca().set_facecolor('#212121')  
plt.gcf().set_facecolor('#212121')  

plt.tick_params(colors='white')

plt.show()
```
