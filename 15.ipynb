## Creating interaction features

```python
import pandas as pd

# Sample dataset
data = {'engine_size': [2.0, 2.5, 3.0], 'fuel_efficiency': [30, 25, 20]}
df = pd.DataFrame(data)

# Creating an interaction feature
df['performance'] = df['engine_size'] * df['fuel_efficiency']
print(df)
```

### Polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures

# Original feature
X = [[2], [3], [4]]

# Creating polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
print(X_poly)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

# Original feature
X = np.array([[2], [3], [4]])

# Creating polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

print(X_poly)
```

```python
# Plotting
plt.figure(figsize=(8, 5))
ax = plt.gca()

# Original feature plot
plt.scatter(X, X_poly[:, 1], color='red', label='Original Feature', zorder=5)

# Polynomial features plot
plt.plot(X, X_poly[:, 1], 'bo-', label='Original Feature')
plt.plot(X, X_poly[:, 2], 'g^--', label='Squared Feature (degree=2)')

# Customizing plot appearance
plt.xlabel('X', color='white')
plt.ylabel('Transformed Features', color='white')
plt.title('Polynomial Features Transformation', color='white')

# Legend with white text and dark background
plt.legend(frameon=True, facecolor='#111111', edgecolor='white', labelspacing=1.2, loc='best')
for text in plt.gca().get_legend().get_texts():
    text.set_color('white')

# Custom background color and no grid
ax.set_facecolor('#111111')
plt.gcf().patch.set_facecolor('#111111')  # For the figure background
plt.grid(False)

# Adjust tick labels and tick colors to be white
ax.tick_params(axis='both', colors='white')

# Ensure equal scaling for x and y axes
ax.set_aspect('auto')  # This ensures that the aspect ratio is not forced to be equal, allowing for correct scaling.

plt.show()
```

### Binning technique

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({'income': [30000, 60000, 90000, 120000]})

# Define the income brackets
df['income_bracket'] = pd.cut(
    df['income'], 
    bins=[0, 50000, 100000, 150000], 
    labels=["low", "medium", "high"])
```
