```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Sample data
data = {
    'X': [1, 2, 3, 4, 5], 
    'Y': [2, 3, 5, 7, 11]
        }
df = pd.DataFrame(data)

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df[['X']], df['Y'], 
    test_size=0.2, 
    random_state=42)



# Creating a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)






```

```python

# Creating a range of values for X 
X_range = pd.DataFrame(np.linspace(df['X'].min(), 
                                   df['X'].max(), 100), 
                                   columns=['X'])

predictions_range = model.predict(X_range)

# Plotting
plt.scatter(df['X'], df['Y'], color='blue', label='Actual Data')
plt.plot(X_range, predictions_range, color='red', label='Predicted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
